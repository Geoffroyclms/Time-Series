```python
df=quandl.get("WIKI/" + "AAPL", start_date=start, end_date=end) 
df=df[['Close']]
df.columns=['AAPL']
# df.to_csv(r"C:\Users\Geoffroy\Desktop\IT\Projet Supélec\Data\Data_brut_AAPL.csv')


for l in logo:
    try:
        df2=quandl.get("WIKI/" + l, start_date=start, end_date=end) 
        df2=df2[['Close']]
        df2.columns=[l]
        # df2.to_csv(r"C:\Users\Geoffroy\Desktop\IT\Projet Supélec\Data\Data_brut_'+l+'.csv"
        df=df.join(df2)
    except:
    #    logo.remove(l)
        pass
    

    
df.head(3)
```

```python
#Enlever les entreprises dont les data sont pas bien (cf NaN)
list_err=df.columns[df.isna().any()].tolist()
for l in list_err:
    logo=logo.drop(index=(logo==l).argmax())
    df=df.drop(columns=l)
    
df.tail(1)
```
